```python
# Week 2 - AI Track
# Data Analysis with Python

import pandas as pd

# 1. Load dataset (Air pollution)
data = pd.read_csv("air_pollution.csv")
print(data)
```

```python
# 2. Display first 5 rows
print("First 5 rows of dataset:")
print(data.head())
```

```python
# 3. Calculate average value of 'Score' column
average_CO2_level = data["CO2_Level"].mean()
print(f"The Average CO2 Level is : {average_CO2_level}")
```

```python
# 4. city that emmit high CO2 gas 
max_CO2_Level  = data["CO2_Level"].max()
max_city = data.loc[data["CO2_Level"].idxmax(), "City"]
```

```python
# 5. Print summary of results
print("\nSummary of Results:")
print(f"Average Score: {average_CO2_level:.2f}")
print(f"Highest CO2_LEVEL: {max_CO2_Level} is emmited by {max_city} city")
```
